### 5. 두 벡터 $a = \left( a_1, a_2, \dots, a_n \right)$, $b = \left( b_1, b_2, \dots, b_n \right)$가 주어졌을 두 벡터간 각도의 코사인 값을 이용하여 두 벡터가 얼마나 유사한지 평가한다. 아래 주어진 수식은 두 벡터 간 각도의 코사인 값 코사인 유사도를 계산하는 식이다.

$$
{\displaystyle {\text{코사인 유사도}}=\cos(\theta )={\mathbf {a} \cdot \mathbf {b}  \over | \mathbf {a} | \cdot | \mathbf {b} |}={\frac {\sum \limits _{i=1}^{n}{a_{i} b_{i}}}{{\sqrt {\sum \limits _{i=1}^{n}{a_{i}^{2}}}}{\sqrt {\sum \limits _{i=1}^{n}{b_{i}^{2}}}}}},}
$$


아래 두 벡터 간의 각도의 코사인 값을 계산하시오.
