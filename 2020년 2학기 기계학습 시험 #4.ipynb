체중과 대사율을 특성(feature, 설명변수)으로 하여 성별(target, 반응변수)로 분류하는 프로그램이다.

```
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import StratifiedKFold, cross_val_score, train_test_split
from sklearn.metrics import confusion_matrix, classification_report

url = "https://raw.githubusercontent.com/joongyang/python-for-raw-beginners/master/datasets/metabolism.csv"
data = pd.read_csv(url)
target = data.gender
features = data[['rate', 'weight']]
```

```
target.head(3)
```

```
#1
target = pd.get_dummies(target)
target = target.iloc[:, 0]
```

```
#2
ss = StandardScaler()
```

```
#3
lr = LogisticRegression()
```

```
#4
pipeline = make_pipeline(ss, lr)
```

```
#5
skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=1234)
```

```
#6
results = cross_val_score(pipeline, features, target,
                          cv=skf, scoring='accuracy')
```

```
#7
pipeline.fit(features, target)
```

```
#8
pred = pipeline.predict(features)
```

```
#9
confusion_matrix(pred, target)
```

```
#10
report = classification_report(pred, target)
print(report)
```
