# 자료와 분석 목적

콘크리트의 수명은 콘크리트의 압축 강도에 의존하고, 콘크리트의 압축 강도는 콘크리트의 구성 성분의 비율에 따라 달라진다.
분석하고자 하는 자료는 1$m^3$에 들어 가는 구성 성분의 양을 달리한 콘크리트의 압축 강도를 측정한 것이다.
이 자료를 이용하여 구성 성분의 양에 따라 압축 강도가 어떻게 변하는지 예측하는 최적 모형을 찾고자 한다.

# Features에 대한 설명

| 변수 | 설명 | 단위 | 비고 |
|:---|:---|:---|:---|
| X1 | 시멘트(cement) | 혼합물 $m^{3}$ 당 kg | |
| X2 | 고로 슬래그 분말(Blast Furnace Slag) | 혼합물 $m^{3}$ 당 kg | |
| X3 | 석회(Fly Ash) | 혼합물 $m^{3}$ 당 kg | |
| X4 | 물(Water) | 혼합물 $m^{3}$ 당 kg | |
| X5 | 감수제(Superplasticizer) | 혼합물 $m^{3}$ 당 kg | 콘크리트 제조에 들어가는 첨가제|
| X6 | 골재(Coarse Aggregate) | 혼합물 $m^{3}$ 당 kg | 굵은 골재(10mm 체를 통과하지 않는 골재) |
| X7 | 잔골재(Fine Aggregate) | 혼합물 $m^{3}$ 당 kg | 작은 골재(10mm 체를 통과하는 골재) |
| X8 | 재령(Age) | Day (1~365) | 콘크리트를 친 후 경과한 일 수|
| Y | 콘크리트 압축 강도 | MPa | Output, Target |
