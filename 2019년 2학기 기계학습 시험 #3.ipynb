# Seed Data
---

Seed Data는 세가지 품종의 밀(Kama, Rosa, Canadian) 씨앗의 모양을 나타내는 여러 feature를 soft X-ray 기법으로 측정한 것이다.
밀 씨앗의 모양을 나타내는 여러 feture로부터 밀의 품종을 알아내는 최적의 기계학습 알고리즘을 찾아내고자 한다.

# Features와 target에 대한 설명
---

밀 씨앗의 모양을 나타내는 특성과 품종은 다음과 같다.

| 칼럼명 | 설명 
|:---|:---|
| X1 | A(area) : 씨앗의 면적($mm^2$)
| X2 | P(perimeter) : 씨앗의 둘레($m$)
| X3 | C(compactness) $\frac{4 \pi A}{P^2}$ : 씨앗의 압축 정도
| X4 | length of kernel : 씨앗의 길이($mm$)
| X5 | width of kernel : 씨앗의 폭($mm$)
| X6 | asymmetry coefficient : 씨앗의 비대칭 계수
| X7 | length of kernel groove : 씨앗에 있는 홈의 길이($mm$)
| Y | 밀의 품종: 1=Kama, 2=Rosa, 3=Canadian
