# Seed Data
---

Seed Data는 세가지 종류의 밀(Kama, Rosa, Canadian)의 알맹이의 내부 구조를 나타내는 여러 feature를 soft X-ray 기법으로 측정한 것이다.
이 자료를 이용하여 내부 구조를 나타내는 feture로부터 밀을 종류를 알아내는 최적의 기계학습 알고리즘을 찾아내고자 한다.

# Features 설명
---

밀 알맹이의 내부 구조를 나타내는 특성

| 칼럼명 | 설명 
|:---|:---|
| X1 | A(area) : 면적($mm^2$)
| X2 | P(perimeter) : 둘레($m$)
| X3 | C(compactness) $\frac{4 \pi A}{P^2}$ : 압축정도
| X4 | length of kernel : 알맹이의 길이($mm$)
| X5 | width of kernel : 알맹이의 폭($mm$)
| X6 | asymmetry coefficient : 비대칭 계수
| X7 | length of kernel groove : 알맹이 홈의 길이($mm$)
| Y | 밀의 종류: 1=Kama, 2=Rosa, 3=Canadian
